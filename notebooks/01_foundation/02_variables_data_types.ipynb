## Core Concepts

**Variable:** Named container for data  
**Data Type:** Kind of data (text, number, true/false)  
**Assignment:** Storing data in variables  
**Conversion:** Changing data from one type to another

## 1. Creating Variables

```python
# Create variables: variable_name = value
name = "Alice"
age = 25
height = 5.6
is_student = True

# Use variables
print(name)
print(f"{name} is {age} years old")

# Variables can change
age = 26
print(f"Next year: {age}")
```

## 2. Data Types

```python
# String (text) - use quotes
name = "Bob"
message = "Hello World"
color = 'blue'

print("Text examples:")
print(f"Name: {name}")
print(f"Message: {message}")
print(f"Color: {color}")

# Integer (whole numbers)
age = 30
score = 100
temperature = -5

print("\nWhole numbers:")
print(f"Age: {age}")
print(f"Score: {score}")
print(f"Temperature: {temperature}")

# Float (decimal numbers)
height = 5.8
price = 19.99
pi = 3.14

print("\nDecimal numbers:")
print(f"Height: {height}")
print(f"Price: ${price}")
print(f"Pi: {pi}")

# Boolean (True/False)
is_sunny = True
is_raining = False
has_license = True

print(f"\nTrue/False values:")
print(f"Sunny today: {is_sunny}")
print(f"Raining: {is_raining}")
print(f"Has license: {has_license}")

# Check data types
print(f"\nData type examples:")
print(f"type(name): {type(name)}")
print(f"type(age): {type(age)}")
print(f"type(height): {type(height)}")
print(f"type(is_sunny): {type(is_sunny)}")
```

## 3. Type Conversion & Operations

```python
# Converting between types
age_text = "25"
age_number = int(age_text)
print(f"Text '{age_text}' ‚Üí Number {age_number}")

price_text = "19.99" 
price_number = float(price_text)
print(f"Text '{price_text}' ‚Üí Number {price_number}")

score = 85
score_text = str(score)
print(f"Number {score} ‚Üí Text '{score_text}'")

# Math with variables
a, b = 10, 3
print(f"\nMath: a={a}, b={b}")
print(f"a + b = {a + b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b:.2f}")

# Text operations
first = "John"
last = "Smith"
full_name = first + " " + last
print(f"\nText: '{first}' + ' ' + '{last}' = '{full_name}'")
print(f"Length: {len(full_name)} characters")
```

## Learning Objectives
- Get user input with `input()`
- Display output with `print()`
- Convert text input to numbers
- Format output professionally
- Build interactive programs

## Core Concepts

**Input:** Getting data from users  
**Output:** Showing results to users  
**Conversion:** Changing text to numbers for math

*Example: Ask user's age ‚Üí Calculate birth year ‚Üí Display result*

```python
# Basic output with print()
print("Hello, World!")
print("Python is awesome!")

# Multiple items and separators
print("Name:", "Alice", "Age:", 25)
print("apple", "banana", "orange", sep=", ")
print("2024", "08", "27", sep="-")

# Formatted output with f-strings
name = "Bob"
age = 30
score = 85.7

print(f"Hello, {name}!")
print(f"You are {age} years old")
print(f"Your score is {score}")

# Number formatting
price = 19.99
print(f"Price: ${price:.2f}")

percentage = 0.75
print(f"Success rate: {percentage:.1%}")
```
